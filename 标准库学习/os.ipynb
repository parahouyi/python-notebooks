```python
# 批量查找指定文件夹下的DOC文档

import os
```

####  walk  生成器，包含的是元组，没有完整目录。元组第一个是目录（字符串），第二个是包含的子文件夹（列表），第三个是文件列表

```python
t = os.walk('e:\\test') 
```

```python
for i in t:
    print(i)
```

```python
files = []

for i in t:
    if i[2]: #表示有文件
        for j in i[2]:
            files.append(i[0] +'\\'+j) #完整目录名加上文件名
```

```python

mp3files = [file for file in files if file.endswith(('doc', 'docx', 'avi'))]
for i in mp3files:
    print(i)
```

```python
def findFiles(path: str):
    """
    find all the files in the path
    :param path: a folder name
    :return:
    """
    all_files = []

    def findAllFiles(path):
        nonlocal all_files
        dirs = os.listdir(path)
        for file in dirs:
            try:
                full_path = os.path.join(path, file)
                if os.path.isfile(full_path):
                    all_files.append(full_path)
                elif os.path.isdir(full_path):
                    findAllFiles(full_path)
            except:
                continue
        return all_files

    return findAllFiles(path)
```

```python
def findAllTypeFiles(path:str, extentions: str | tuple = None) -> list:
    """
    get the list of give type files in the folder path
    :param path: folder name
    :param extentions: file type string or tuple of strings [lowercase]
    :return:
    """
    all_files = findFiles(path)
    if extentions is not None:
        all_type_files = []
        for file in all_files:
            if file.endswith(extentions):
                all_type_files.append(file)
        # all_type_files = [file for file in all_files if file.endswith(extentions)]
        return all_type_files
    return all_files
```
