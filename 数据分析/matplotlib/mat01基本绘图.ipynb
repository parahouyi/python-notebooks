````python
x = np.linspace(-np.pi, np.pi, 1000)
sinx = np.sin(x)
# 画一个y = cos(x)/2
cosx = np.cos(x) / 2
````

<pre># 设置图例的位置
# loc: <关键字参数> 制定图例的显示位置 (若不设置loc，则显示默认位置)

````python
plt.plot(x, sinx, label='sin(x)',color=(0.8,0.5,0.5,0.8),linewidth=3)
# plt.xlim(-1,1)
# plt.ylim(-0.5,0.5)
plt.plot(x,cosx,label='cos(x)',color='orangered',linestyle='-.')
plt.legend(loc='lower right')
vals = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
texts = [r'$-\pi$', r'$-\frac{\pi}{2}$', '0', 
		 r'$\frac{\pi}{2}$', r'$\pi$']
plt.xticks(vals, texts)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
plt.yticks([-1.0, -0.5, 0.5, 1.0])
plt.scatter([np.pi/2,np.pi/2],[1,0],marker='h',s=[170,50],color=['b','r'],zorder=3)
# 为特殊点添加备注
plt.annotate(r'$[\frac{\pi}{2}, 1]$', 
	xycoords='data', xy=(np.pi/2, 1), 
	textcoords='offset points', xytext=(10, 20),
	fontsize=14, arrowprops=dict(
		arrowstyle='->',
		connectionstyle='angle3'))
````

````python
# 设置坐标轴
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
plt.yticks([-1.0, -0.5, 0.5, 1.0])
````
